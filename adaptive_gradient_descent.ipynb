```python
import numpy as np
import matplotlib.pyplot as plt
from math import *


def gradient_descent(x, y, f, df, iterations=1000, learning_rate=0.0001, stopping_threshold=1e-6,costs=[]):
    
    # Initializing weight, bias, learning rate, and iterations
    current_x = np.array([x, y])
    iterations = iterations
    learning_rate = learning_rate
    n = float(len(current_x))
    
    xs = []
    previous_cost = None

    # Estimation of optimal parameters
    for i in range(iterations):
        # Making predictions
        y_predicted = f(*current_x)

        # Calculating the current cost
        current_cost = y_predicted

        # If the change in cost is less than or equal to stopping_threshold, we stop the gradient descent
        if previous_cost and abs(previous_cost - current_cost) <= stopping_threshold:
            break

        previous_cost = current_cost

        costs.append(current_cost)
        xs.append(current_x.copy())

        # Calculating the gradients
        gradient = df(*current_x)

        # Updating x
        current_x = np.array(current_x) - learning_rate * gradient

        # Printing the parameters for each 100th iteration
        if (i + 1) % 100 == 0:
            print(f"Iteration {i+1}: Cost {current_cost}, x {current_x}")

    # Plotting the path of x during optimization
    xs = np.array(xs)
    plt.figure(figsize=(8, 6))
    plt.plot(xs[:, 0], xs[:, 1], 'bo-', alpha=0.3, label=None)
    plt.title("Optimization Path")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.show()

    # Plotting the cost vs iterations
    # plt.figure(figsize=(8, 6))
    # plt.plot(costs)
    # plt.title("Cost vs Iterations")
    # plt.ylabel("Cost")
    # plt.xlabel("Iterations")
    # plt.show()
```

```python
# implementation of functions to test and compare results with the new algorithmic further developed ( Gradient descent without descent)
from math import *
# three hump camel function
def testFunction1(x,y) :
  return 2*x**2 - 1.05*x**4 + x**6/6 + x*y + y**2


def mySquareRoot( number):
  return math.sqrt(number)

def calculateGradientFunction1(x, y):
    # Partial derivatives of the function
    dx = 4*x - 4.2*x**3 + x**5
    dy = x + 2*y

    return np.array([dx, dy])

#eason fucntion
def testFunction2(x,y) :
  return -np.cos(x) * np.cos(y) * np.e **( - ((x-np.pi) ** 2 + (y - np.pi) **2))



def calculateGradient2(x, y):
    # Partial derivatives of the function
    dx = np.cos(x) * np.cos(y) * np.e ** ( - ((x - np.pi) ** 2 + (y - np.pi) ** 2)) * (2 * (x - np.pi))
    dy = np.cos(x) * np.cos(y) * np.e ** ( - ((x - np.pi) ** 2 + (y - np.pi) ** 2)) * (2 * (y - np.pi))

    return np.array([dx, dy])

#egg holder function
def testFunction3(x,y):
 return (-(y + 47) * np.sin(np.sqrt(np.abs(x/2 + (y + 47)))) -x * np.sin(np.sqrt(np.abs(x - (y + 47)))))

def gradient_testFunction3(x, y):
    # Calculate partial derivatives using numerical approximation

    # Step size for numerical differentiation
    h = 1e-6

    # Partial derivative with respect to x
    df_dx = (testFunction3(x + h, y) - testFunction3(x - h, y)) / (2 * h)

    # Partial derivative with respect to y
    df_dy = (testFunction3(x, y + h) - testFunction3(x, y - h)) / (2 * h)

    # Return the gradient as a tuple
    return np.array([df_dx , df_dy])

#plot function method
def plotAFunction(func, x_range, y_range):
    x = np.linspace(x_range[0], x_range[1], 100)
    y = np.linspace(y_range[0], y_range[1], 100)
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)

    # Plot 3D representation
    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='hot')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # Plot contour lines
    plt.figure(figsize=(8, 6))
    plt.contourf(X, Y, Z, cmap='hot')
    plt.colorbar()
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Contour Lines')

    plt.show()


```
